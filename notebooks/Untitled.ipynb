```python
# Read Text Files with Pandas using read_csv()
  
# importing pandas
from matplotlib import pyplot as plt 
import pandas as pd

plt.rcParams.update({'font.size': 18})
```

```python
# read text file into pandas DataFrame
kept = pd.read_csv("/gpfs/home/asun/nanopore/20230906_lam/basecalling/pass/barcode12/internalKeepTrimmed.txt", sep="\t", header=None)
discarded = pd.read_csv("/gpfs/home/asun/nanopore/20230906_lam/basecalling/pass/barcode12/internalDiscardTrimmed.txt", sep="\t", header=None)

# display DataFrame
#display(df)

keptReadLength = kept[0].to_numpy()
discardedReadLength = discarded[0].to_numpy()

# Creating histogram
fig, ax = plt.subplots(figsize =(15, 10))
n, bins, patches = ax.hist(keptReadLength, bins=50, label="Reads with Internal Adapters", alpha=0.6)
ax.hist(discardedReadLength, bins=bins, label="Reads without Internal Adapters", alpha=0.6)
ax.legend()
ax.set_xlabel("Read Length (BP)")
ax.set_ylabel("Number of Reads")
plt.title(f'Barcode 12 Read Length Distribution')
#plt.xticks(range(0, 7500, 500))
plt.savefig(f'bc12canonReadDist.png', facecolor='white')

# Show plot
plt.show()
```
