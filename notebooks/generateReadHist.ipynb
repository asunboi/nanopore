```python
# Read Text Files with Pandas using read_csv()
  
# importing pandas
from matplotlib import pyplot as plt 
import pandas as pd

plt.rcParams.update({'font.size': 14})
```

```python
import os
os.path.join(os.path.dirname(__file__))
my_path = os.path.abspath(__file__)
    
# read text file into pandas DataFrame
df = pd.read_csv("~/nanopore/20230906_lam/basecalling/pass/barcode10/readLength", sep="\t", header=None)
  
# display DataFrame
# display(df)

readLength = df[2].to_numpy()
 
# Creating histogram
fig, ax = plt.subplots(figsize =(15, 10))
ax.hist(readLength, bins=50)
ax.set_xlabel("Read Length (BP)")
ax.set_ylabel("Number of Reads")
plt.title('KAPA Read Length vs Number of Reads')
plt.xticks(range(0, 7500, 500))
plt.savefig('KAPA.png')

# Show plot
plt.show()
```
