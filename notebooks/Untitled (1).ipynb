```python
# Read Text Files with Pandas using read_csv()
  
# importing pandas
from matplotlib import pyplot as plt 
import pandas as pd

plt.rcParams.update({'font.size': 18})
```

```python
# read text file into pandas DataFrame
pool = pd.read_csv("poolReadLengths.txt", sep="\t", header=None)
trimmed = pd.read_csv("trimmedReadLengths.txt", sep="\t", header=None)

# display DataFrame
#display(df)

poolReadLength = pool[0].to_numpy()
trimmedReadLength = trimmed[0].to_numpy()

# Creating histogram
fig, ax = plt.subplots(figsize =(15, 10))
n, bins, patches = ax.hist(poolReadLength, bins=50, label="New Processing", alpha = 0.6)
ax.hist(trimmedReadLength, bins=bins, label="Original Processing", alpha = 0.6)
ax.legend()
ax.set_xlabel("Read Length (BP)")
ax.set_ylabel("Number of Reads")
plt.title(f'Barcode 4 Read Length Distribution')
#plt.xticks(range(0, 7500, 500))
plt.savefig(f'bc4ReadDist.png', facecolor='white')

# Show plot
plt.show()
```
